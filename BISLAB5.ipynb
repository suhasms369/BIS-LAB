```python
import numpy as np

def evaluate_fitness(schedule, job_times):
    total_time = 0
    for task in schedule:
        total_time += job_times[task]
    return total_time

def levy_flight(schedule):
    new_schedule = schedule.copy()
    i, j = np.random.choice(len(new_schedule), 2, replace=False)
    new_schedule[i], new_schedule[j] = new_schedule[j], new_schedule[i]
    return new_schedule

def cuckoo_search(job_times, num_nests, max_iterations):
    nests = [np.random.permutation(len(job_times)) for _ in range(num_nests)]
    best_schedule = None
    best_fitness = float('inf')

    for iteration in range(max_iterations):
        for i in range(num_nests):
            new_schedule = levy_flight(nests[i])
            new_fitness = evaluate_fitness(new_schedule, job_times)

            if new_fitness < evaluate_fitness(nests[i], job_times):
                nests[i] = new_schedule

            if new_fitness < best_fitness:
                best_fitness = new_fitness
                best_schedule = new_schedule

        abandon_indices = np.random.choice(num_nests, int(num_nests * 0.25), replace=False)
        for index in abandon_indices:
            nests[index] = np.random.permutation(len(job_times))

    return best_schedule, best_fitness

job_times = [3, 2, 5, 4, 1]

best_schedule, best_fitness = cuckoo_search(job_times, num_nests=11, max_iterations=500)

print("Best Schedule:", best_schedule)
print("Best Total Time:", best_fitness)
```
